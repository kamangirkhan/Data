```python
import pandas as pd
import matplotlib.pyplot as plt

#Load data
tips_df = pd.read_csv('tips.csv')
tips_df

# Do smoker people give more tips + Pie chart
## find the total amount of smokers + non smokers
both_smoker = len(tips_df['smoker'])
yes_smoker = len(tips_df[tips_df['smoker'] == 'Yes'])
no_smoker = len(tips_df[tips_df['smoker'] == 'No'])
yes_smoker_percent = yes_smoker / both_smoker
no_smoker_percent = no_smoker / both_smoker

## Pie chart
fig, ax = plt.subplots()
ax.set_title('Proportion of tips by smoker condition')
ax.pie([yes_smoker_percent, no_smoker_percent], labels=['Yes', 'No'], autopct='%1.2f%%')
ax.axis('equal')
ax.set_ylabel('Is smoker?')
plt.show()
print()
```

```python
# Chart 1: Scatter plot - total_bill vs tip
plt.figure()
plt.scatter(tips_df["total_bill"], tips_df["tip"])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.show()
```

```python
# Chart 3: Boxplot - tip by sex
plt.figure()
tips_df.boxplot(column="tip", by="sex")
plt.title("Tip Distribution by Sex")
plt.suptitle("")  # removes the automatic extra title
plt.xlabel("Sex")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.show()
```

```python
# Chart 5: Bar chart - average total_bill by time (Lunch/Dinner)
avg_bill_by_time = tips_df.groupby("time")["total_bill"].mean()
plt.figure()
avg_bill_by_time.plot(kind="bar")
plt.title("Average Total Bill by Time of Day")
plt.xlabel("Time")
plt.ylabel("Average Total Bill ($)")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.show()
```

```python
# Extra Chart 3: Line chart - average total bill by party size
avg_bill_by_size = tips_df.groupby("size")["total_bill"].mean()
plt.figure()
plt.plot(avg_bill_by_size.index, avg_bill_by_size.values, marker="o")
plt.title("Average Total Bill by Party Size")
plt.xlabel("Party Size")
plt.ylabel("Average Total Bill ($)")
plt.grid(True)
plt.show()
```
